```python
nums = list(range(0,1449))
```

```python
for i, num in enumerate(nums):
    # print(num)
    num = str(num)
    # print(num)
    if len(num) < 4:
        total_len = len(num)
        while total_len < 4:
            num = '0' + num
            # print(num)
            total_len += 1
    nums[i] = num
```

```python
print(nums)
```

```python
import os
```

```python
import random

# Function to split the list into subsets
def split_list(input_list, proportions, seed = 42):
    random.seed(10)

    # Shuffle the list randomly
    shuffled_list = input_list.copy() 
    random.shuffle(shuffled_list)
    
    # Calculate the lengths of subsets based on proportions
    total_length = len(shuffled_list)
    lengths = [int(total_length * proportion) for proportion in proportions]
    
    # Split the list into subsets
    subsets = []
    start = 0
    for length in lengths:
        subsets.append(shuffled_list[start:start+length])
        start += length
    
    return subsets
```

```python
# Getting the random split

# Proportions of 70%, 20%, and 10%
proportions = [0.7, 0.2, 0.1]

# Split the list
subsets = split_list(nums, proportions, seed=10)

# Print the subsets
for i, subset in enumerate(subsets, 1):
    print(f"Subset {i}: {subset}")
```

```python
train = subsets[0]
val = subsets[1]
test = subsets[2]
```

```python
subsets[0]
```

```python
data_filled = {}
sets = ['training', 'validation', 'testing']
headers = ['train', 'val', 'test']

for i, set_name in enumerate(sets):
    header = headers[i]
    print(header)

    files = []
    for j in subsets[i]:
        img = set_name + '/img/' + j + '.png'
        depth = set_name + '/depth/' + j + '.npy'
        seg = set_name + '/seg/' + j + '.png'
        # seg_vis = set_name + '/seg_vis/' + j + '.png'
        normal_mask = set_name + '/normal_mask/' + j + '.png'

        files.append([img, seg, normal_mask, depth])
    
    data_filled[header] = files
```

```python
len(data_filled['test'])
```

```python
# Path to the JSON file
file_path = "new_data/nyu_v2/nyu_v2_3task.json"

# Write data to the JSON file
with open(file_path, "w") as json_file:
    json.dump(data_filled, json_file)
```

```python
locations = ['depth', 'img', 'normal_mask', 'seg', 'seg_vis']
for num in nums:
    if num in train:
        continue
    elif num in val:
        source = "new_data/nyu_v2/training"
        target = "new_data/nyu_v2/validation"
        for location in locations:
            if location != 'depth':
                file_name = str(num) + ".png"
            else:
                file_name = str(num) + ".npy"
            full_source = os.path.join(source, location, file_name)
            full_target = os.path.join(target, location, file_name)
            try:
                # Move the file using os.rename()
                os.rename(full_source, full_target)
                print(f"File '{full_source}' moved to '{full_target}'")
            except FileNotFoundError:
                print(f"Error: File '{full_source}' not found.")
    elif num in test:
        source = "new_data/nyu_v2/training"
        target = "new_data/nyu_v2/testing"
        for location in locations:
            if location != 'depth':
                file_name = str(num) + ".png"
            else:
                file_name = str(num) + ".npy"
            full_source = os.path.join(source, location, file_name)
            full_target = os.path.join(target, location, file_name)
            try:
                # Move the file using os.rename()
                os.rename(full_source, full_target)
                print(f"File '{full_source}' moved to '{full_target}'")
            except FileNotFoundError:
                print(f"Error: File '{full_source}' not found.")
    else:
        print(num)
```
